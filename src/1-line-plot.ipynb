### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate x values ranging from 0 to 10
x = np.arange(0, 11)

# Generate y values as the square of x
y = x**2
```

```python

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add title to the plot
plt.title('Line Plot of y = x^2')

# Add labels to the axes
plt.xlabel('x values')
plt.ylabel('y values (x^2)')

# Add a grid to the plot
plt.grid(True)

# Show the plot
plt.show()
```
